<br><br><strong>Cách hoạt động:</strong>

<strong>Thuật toán <code>Minimax</code> trong tìm kiếm đối kháng:</strong>
Minimax là thuật toán cốt lõi trong AI cho các trò chơi hai người (như Tic-Tac-Toe, Cờ vua, Caro). Thuật toán mô phỏng toàn bộ các nước đi có thể xảy ra, giả định cả hai bên đều chơi hoàn hảo, từ đó chọn ra hành động tối ưu.
<br><br><strong>Ý tưởng chính:</strong>

Một người chơi cố gắng <strong>tối đa hóa</strong> điểm số (Max – thường là người chơi X).

Người còn lại cố gắng <strong>tối thiểu hóa</strong> điểm số (Min – thường là người chơi O).

Thuật toán duyệt toàn bộ cây trạng thái để tìm hành động đảm bảo kết quả tốt nhất ngay cả khi đối thủ chơi tối ưu.
<br><br><strong>Cách hoạt động:</strong>

<ol> <li><strong>Xét trạng thái hiện tại:</strong> Thuật toán xác định đang tới lượt Max hay Min.</li> <li><strong>Duyệt tất cả các hành động hợp lệ:</strong> Mỗi hành động dẫn tới một trạng thái mới.</li> <li><strong>Gọi đệ quy để mô phỏng nước đi tiếp theo của đối thủ:</strong> <ul> <li>Nếu đang là Max → gọi <code>minValue()</code> để mô phỏng nước đi của Min.</li> <li>Nếu đang là Min → gọi <code>maxValue()</code> để mô phỏng nước đi của Max.</li> </ul> </li> <li><strong>Dừng lại khi đến trạng thái kết thúc:</strong> Nếu trạng thái là thắng/thua/hòa → trả về giá trị <code>utility</code>: <ul> <li>X thắng → <code>+1</code></li> <li>O thắng → <code>-1</code></li> <li>Hòa → <code>0</code></li> </ul> </li> <li><strong>Chọn giá trị tối ưu:</strong> <ul> <li>Max chọn giá trị <strong>lớn nhất</strong>.</li> <li>Min chọn giá trị <strong>nhỏ nhất</strong>.</li> </ul> </li> </ol>
<br><strong>Kết quả:</strong>
Minimax đảm bảo rằng người chơi sẽ đưa ra nước đi tốt nhất có thể, ngay cả trong tình huống đối thủ luôn chơi “hoàn hảo”.

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 10px; border-radius: 8px; max-width: 600px;">
  <strong>Hàm <code>initial_state()</code>:</strong> Khởi tạo và trả về <em>trạng thái ban đầu của bàn cờ</em> trong trò chơi Tic Tac Toe.
  <br><strong>Cách hoạt động:</strong> Hàm tạo một ma trận 3×3, trong đó tất cả các ô đều mang giá trị <code>EMPTY</code>, thể hiện rằng bàn cờ hoàn toàn trống và chưa có người chơi nào đi nước đầu tiên.
  <br><strong>Giá trị trả về:</strong> Một danh sách 2 chiều 3×3 gồm toàn bộ các ô trống <code>EMPTY</code>.
</div>


```python
def initial_state():
  return [[EMPTY, EMPTY, EMPTY],
          [EMPTY, EMPTY, EMPTY],
          [EMPTY, EMPTY, EMPTY]]
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 10px; border-radius: 8px; max-width: 650px;">
  <strong>Hàm <code>player(board)</code>:</strong> Xác định <em>người chơi sẽ đi tiếp theo</em> dựa trên số lượng ô đã được đánh trên bàn cờ.
  <br><strong>Cách hoạt động:</strong> Hàm duyệt toàn bộ các ô của bàn cờ và đếm số ô không rỗng. Nếu tổng số ô đã được đánh là số lẻ, điều đó nghĩa là AI vừa đi xong và đến lượt người dùng; nếu là số chẵn thì đến lượt AI.
  <br><strong>Giá trị trả về:</strong> Trả về <code>ai</code> hoặc <code>user</code> tùy theo lượt tương ứng.

```python
def player(board):
  count = 0
  for i in board:
    for j in i:
      if j:
        count += 1
  if count % 2 != 0:
    return ai
  return user
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 10px; border-radius: 8px; max-width: 600px;">
  <strong>Hàm <code>actions(board)</code>:</strong> Trả về <em>tập hợp các nước đi hợp lệ</em> trên bàn cờ (các ô trống).
  <br><strong>Tham số:</strong> <code>board</code> – danh sách 2 chiều, mỗi phần tử là <code>EMPTY</code> hoặc ký hiệu người chơi.
  <br><strong>Cách hoạt động:</strong> Duyệt tất cả các ô, nếu ô trống (<code>board[i][j] == EMPTY</code>) thì thêm tọa độ <code>(i,j)</code> vào tập kết quả.
  <br><strong>Giá trị trả về:</strong> Set các tuple <code>(i, j)</code> biểu diễn các ô trống.
</div>


```python
def actions(board):
  res = set()
  board_len = len(board)
  for i in range(board_len):
    for j in range(board_len):
      if board[i][j] == EMPTY:
        res.add((i, j))
  return res
```

<div style="font-family: Arial, sans-serif; border: 2px solid #2196F3; padding: 10px; border-radius: 8px; max-width: 600px;">
  <strong>Hàm <code>result(board, action)</code>:</strong> Trả về bàn cờ mới sau khi người chơi hiện tại thực hiện nước đi <code>action</code>.
  <br><strong>Tham số:</strong>
  <ul style="margin:0; padding-left:20px;">
    <li><code>board</code>: danh sách 2 chiều, trạng thái hiện tại của bàn cờ.</li>
    <li><code>action</code>: tuple <code>(i, j)</code>, tọa độ ô muốn đánh.</li>
  </ul>
  <strong>Cách hoạt động:</strong>
  <ol style="margin:0; padding-left:20px;">
    <li>Xác định người chơi hiện tại <code>curr_player = player(board)</code>.</li>
    <li>Tạo bản sao sâu của bàn cờ để không thay đổi bàn gốc.</li>
    <li>Đánh dấu nước đi: <code>result_board[i][j] = curr_player</code>.</li>
    <li>Trả về bàn cờ mới.</li>
  </ol>
  <strong>Giá trị trả về:</strong> Bàn cờ mới (list 2 chiều) sau khi thực hiện nước đi.
</div>


```python
def result(board, action):
  curr_player = player(board)
  result_board = copy.deepcopy(board)
  (i, j) = action
  result_board[i][j] = curr_player
  return result_board
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>get_horizontal_winner(board)</code>:</strong> Kiểm tra xem có hàng nào trên bàn cờ mà tất cả ô đều giống nhau, xác định người thắng theo hàng ngang. <br><strong>Cách hoạt động:</strong> Duyệt từng hàng của bàn cờ. Với mỗi hàng, lấy giá trị của ô đầu tiên làm <em>ứng viên thắng</em>. Nếu tất cả các ô trong hàng đều bằng giá trị này và không phải là <code>None</code>, hàm trả về giá trị đó (người thắng). Nếu không có hàng nào thỏa mãn, trả về <code>None</code>. <br><strong>Giá trị trả về:</strong> <code>'X'</code>, <code>'O'</code> nếu có hàng thắng, <code>None</code> nếu không có hàng thắng. <br> </div>


```python
def get_horizontal_winner(board):
  winner_val = None
  board_len = len(board)
  for i in range(board_len):
    winner_val = board[i][0]
    for j in range(board_len):
      if board[i][j] != winner_val:
        winner_val = None
    if winner_val:
      return winner_val
  return winner_val
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>get_vertical_winner(board)</code>:</strong> Kiểm tra xem có cột nào trên bàn cờ mà tất cả ô đều giống nhau, xác định người thắng theo cột dọc. <br><strong>Cách hoạt động:</strong> Duyệt từng cột của bàn cờ. Với mỗi cột, lấy giá trị của ô đầu tiên làm <em>ứng viên thắng</em>. Nếu tất cả các ô trong cột đều bằng giá trị này và không phải là <code>None</code>, hàm trả về giá trị đó (người thắng). Nếu không có cột nào thỏa mãn, trả về <code>None</code>. <br><strong>Giá trị trả về:</strong> <code>'X'</code>, <code>'O'</code> nếu có cột thắng, <code>None</code> nếu không có cột thắng. </div>

```python
def get_vertical_winner(board):
  winner_val = None
  board_len = len(board)
  for i in range(board_len):
    winner_val = board[0][i]
    for j in range(board_len):
      if board[j][i] != winner_val:
        winner_val = None
    if winner_val:
      return winner_val
  return winner_val
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>get_diagonal_winner(board)</code>:</strong> Kiểm tra xem có đường chéo nào trên bàn cờ mà tất cả ô đều giống nhau, xác định người thắng theo đường chéo. <br><strong>Cách hoạt động:</strong>
<ol>
<li>Kiểm tra đường chéo chính (từ góc trên trái xuống góc dưới phải). Lấy giá trị của ô đầu tiên làm <em>ứng viên thắng</em>. Nếu tất cả các ô trên đường chéo đều bằng giá trị này và không phải <code>None</code>, hàm trả về giá trị đó.</li>
<li>Nếu đường chéo chính không có người thắng, kiểm tra đường chéo phụ (từ góc trên phải xuống góc dưới trái) theo cùng cách.</li>
<li>Nếu không có đường chéo nào thỏa mãn, hàm trả về <code>None</code>.</li>
</ol>
<strong>Giá trị trả về:</strong> <code>'X'</code>, <code>'O'</code> nếu có đường chéo thắng, <code>None</code> nếu không có đường chéo thắng. </div>

```python
def get_diagonal_winner(board):
  winner_val = None
  board_len = len(board)
  winner_val = board[0][0]
  for i in range(board_len):
    if board[i][i] != winner_val:
      winner_val = None
  if winner_val:
    return winner_val
  winner_val = board[0][board_len - 1]
  for i in range(board_len):
    j = board_len - 1 - i
    if board[i][j] != winner_val:
      winner_val = None
  return winner_val
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>winner(board)</code>:</strong> Xác định người thắng trên bàn cờ bằng cách kết hợp kiểm tra hàng ngang, cột dọc và đường chéo. <br><strong>Cách hoạt động:</strong>
<ol>
<li>Gọi lần lượt các hàm <code>get_horizontal_winner(board)</code>, <code>get_vertical_winner(board)</code> và <code>get_diagonal_winner(board)</code>.</li>
<li>Hàm trả về giá trị người thắng đầu tiên được tìm thấy (theo thứ tự: hàng → cột → đường chéo).</li>
<li>Nếu không có người thắng, hàm trả về <code>None</code>.</li>
</ol>
<strong>Giá trị trả về:</strong> <code>'X'</code>, <code>'O'</code> nếu có người thắng, <code>None</code> nếu chưa có người thắng. </div>


```python
def winner(board):
  winner_val = get_horizontal_winner(board) or get_vertical_winner(board) or get_diagonal_winner(board) or None
  return winner_val
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>utility(board)</code>:</strong> Đánh giá giá trị của trạng thái bàn cờ cho thuật toán Minimax. <br><strong>Cách hoạt động:</strong>
<ol>
<li>Gọi hàm <code>winner(board)</code> để xác định người thắng (nếu có).</li>
<li>Nếu người thắng là <code>X</code>, trả về <code>1</code> (có lợi cho X).</li>
<li>Nếu người thắng là <code>O</code>, trả về <code>-1</code> (có lợi cho O).</li>
<li>Nếu không có người thắng, hàm không trả về giá trị nào (có thể coi là <code>None</code>).</li>
</ol>
<strong>Giá trị trả về:</strong> <code>1</code> nếu X thắng, <code>-1</code> nếu O thắng, <code>None</code> nếu chưa có người thắng. </div>


```python
def utility(board):
  winner_val = winner(board)
  if winner_val == X:
    return 1
  elif winner_val == O:
    return -1
  return 0
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>terminal(board)</code>:</strong> Xác định xem trạng thái bàn cờ hiện tại có phải là trạng thái kết thúc hay không. <br><strong>Cách hoạt động:</strong>
<ol>
<li>Gọi hàm <code>winner(board)</code>. Nếu có người thắng, trả về <code>True</code> (trò chơi kết thúc).</li>
<li>Kiểm tra tất cả các ô trên bàn cờ. Nếu còn bất kỳ ô nào trống (<code>EMPTY</code>), trò chơi chưa kết thúc → trả về <code>False</code>.</li>
<li>Nếu không còn ô trống và không có người thắng, trò chơi kết thúc hòa → trả về <code>True</code>.</li>
</ol>
<strong>Giá trị trả về:</strong> <code>True</code> nếu bàn cờ ở trạng thái kết thúc (thắng hoặc hòa), <code>False</code> nếu trò chơi vẫn tiếp tục. </div>


```python
def terminal(board):
  if winner(board) != None:
    return True
  for i in board:
    for j in i:
      if j == EMPTY:
        return False
  return True
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>maxValue(state)</code>:</strong> Tính giá trị tối đa mà người chơi X có thể đạt được từ trạng thái bàn cờ hiện tại theo thuật toán Minimax. <br><strong>Cách hoạt động:</strong>
<ol>
<li>Kiểm tra nếu trạng thái bàn cờ là <code>terminal</code> (kết thúc), trả về giá trị <code>utility(state)</code>.</li>
<li>Khởi tạo giá trị tối đa <code>v</code> với <code>-∞</code>.</li>
<li>Duyệt tất cả các <code>action</code> hợp lệ từ trạng thái hiện tại.</li>
<li>Gọi đệ quy <code>minValue(result(state, action))</code> để tính giá trị phản ứng tốt nhất của đối thủ (O) và cập nhật <code>v = max(v, ...)</code>.</li>
<li>Trả về <code>v</code>, giá trị tối ưu mà X có thể đạt được.</li>
</ol>
<strong>Giá trị trả về:</strong> Số thực đại diện cho giá trị tối ưu cho người chơi X từ trạng thái <code>state</code>. </div>


```python
def maxValue(state):
  if terminal(state):
    return utility(state)
  v = -math.inf
  for action in actions(state):
    v = max(v, minValue(result(state, action)))
  return v
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;"> <strong>Hàm <code>minValue(state)</code>:</strong> Tính giá trị tối thiểu mà người chơi O có thể đạt được từ trạng thái bàn cờ hiện tại theo thuật toán Minimax. <br><strong>Cách hoạt động:</strong> <ol> <li>Kiểm tra nếu trạng thái bàn cờ là <code>terminal</code> (kết thúc), trả về giá trị <code>utility(state)</code>.</li> <li>Khởi tạo giá trị tối thiểu <code>v</code> với <code>∞</code>.</li> <li>Duyệt tất cả các <code>action</code> hợp lệ từ trạng thái hiện tại.</li> <li>Gọi đệ quy <code>maxValue(result(state, action))</code> để tính giá trị phản ứng tốt nhất của đối thủ (X) và cập nhật <code>v = min(v, ...)</code>.</li> <li>Trả về <code>v</code>, giá trị tối ưu mà O có thể đạt được.</li> </ol> <strong>Giá trị trả về:</strong> Số thực đại diện cho giá trị tối ưu cho người chơi O từ trạng thái <code>state</code>. </div>


```python
def minValue(state):
  if terminal(state):
    return utility(state)
  v = math.inf
  for action in actions(state):
    v = min(v, maxValue(result(state, action)))
  return v
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>minimax(board)</code>:</strong>
Xác định nước đi tối ưu cho người chơi hiện tại bằng thuật toán Minimax cơ bản (không Alpha–Beta). Hàm dựa trên việc gọi trực tiếp <code>maxValue</code> hoặc <code>minValue</code> để đánh giá từng nước đi.

<br><strong>Cách hoạt động:</strong>

<ol>
<li>Xác định người chơi hiện tại bằng <code>player(board)</code>.</li>

<li>Nếu là lượt của X:
    <ul>
        <li>Khởi tạo giá trị tốt nhất với <code>-∞</code>.</li>
        <li>Duyệt qua tất cả các <code>action</code> hợp lệ.</li>
        <li>Gọi <code>minValue(result(board, action))</code> để xem đối thủ (O) sẽ phản ứng thế nào.</li>
        <li>Nếu giá trị trả về lớn hơn giá trị hiện tại, cập nhật nước đi tối ưu.</li>
    </ul>
</li>

<li>Nếu là lượt của O:
    <ul>
        <li>Khởi tạo giá trị tốt nhất với <code>+∞</code>.</li>
        <li>Duyệt các hành động hợp lệ.</li>
        <li>Gọi <code>maxValue(result(board, action))</code> để xem phản ứng tối ưu của X.</li>
        <li>Chọn nước đi mang lại giá trị nhỏ nhất (tối ưu cho O).</li>
    </ul>
</li>

<li>Sau khi duyệt xong tất cả các nước đi, trả về <code>move</code> — nước đi tối ưu cho người chơi hiện tại.</li>
</ol>

<strong>Giá trị trả về:</strong>
Một tuple tọa độ <code>(i, j)</code> biểu thị nước đi tốt nhất theo Minimax cơ bản.</div>


```python
def minimax(board):
  current_player = player(board)
  if current_player == X:
    min = -math.inf
    for action in actions(board):
      check = minValue(result(board, action))
      if check > min:
        min = check
        move = action
  else:
    max = math.inf
    for action in actions(board):
      check = maxValue(result(board, action))
      if check < max:
        max = check
        move = action
  return move
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 750px;">

<strong>Đoạn code vòng lặp điều khiển trò chơi:</strong>
Phần này quản lý toàn bộ luồng chơi Tic-Tac-Toe giữa người và AI sử dụng thuật toán Minimax.

<br><br><strong>Cách hoạt động:</strong>

<ol> <li><strong>Khởi tạo trò chơi:</strong> <ul> <li>Tạo bàn cờ mới bằng <code>initial_state()</code>.</li> <li>Đặt biến <code>ai_turn = False</code> để AI chưa đánh ngay ở lượt đầu.</li> <li>Người chơi nhập lựa chọn X hoặc O.</li> <li>Gán ký hiệu của AI dựa trên lựa chọn của người chơi.</li> </ul> </li> <li><strong>Vòng lặp chính của trò chơi:</strong> Chạy liên tục cho đến khi ván đấu kết thúc.</li> <li><strong>Kiểm tra trạng thái ván đấu:</strong> <ul> <li>Dùng <code>terminal(board)</code> để xem ván đấu đã kết thúc chưa.</li> <li>Nếu kết thúc: <ul> <li>Lấy người thắng bằng <code>winner(board)</code>.</li> <li>In ra kết quả thắng/thua/hòa.</li> <li><strong>Thoát vòng lặp.</strong></li> </ul> </li> </ul> </li> <li><strong>Xác định người chơi hiện tại:</strong> <ul> <li>Dùng <code>player(board)</code> để xác định lượt X hay O.</li> </ul> </li> <li><strong>Xử lý lượt của AI:</strong> <ul> <li>Điều kiện: <code>user != playr</code> nghĩa là lượt này không phải của bạn.</li> <li>Nếu <code>ai_turn == True</code>: <ul> <li>AI chọn nước đi tối ưu bằng <code>minimax(board)</code>.</li> <li>Cập nhật bàn cờ: <code>board = result(board, move)</code>.</li> <li>Đặt <code>ai_turn = False</code> để chuyển lượt về người.</li> <li>In bàn cờ bằng <code>numpy.array(board)</code>.</li> </ul> </li> </ul> </li> <li><strong>Xử lý lượt của người chơi:</strong> <ul> <li>Điều kiện: <code>user == playr</code>.</li> <li>Đặt <code>ai_turn = True</code> để AI sẽ đánh ở lượt kế tiếp.</li> <li>Hỏi người chơi nhập vị trí (row, col).</li> <li>Nếu ô được chọn đang trống: <ul> <li>Cập nhật bàn cờ bằng <code>result(board, (i, j))</code>.</li> <li>In bàn cờ.</li> </ul> </li> </ul> </li> </ol>

<strong>Kết quả:</strong>
Đoạn mã này điều khiển toàn bộ trò chơi giữa người và AI, giúp nước đi luân phiên đúng thứ tự, đồng thời để AI sử dụng Minimax chọn nước đi tối ưu.

</div>

```python
board = initial_state()
ai_turn = False
print('Choose a player')
user = input()
if user == 'X':
  ai = 'O'
else:
  ai = 'X'
while True:
  game_over = terminal(board)
  playr = player(board)
  if game_over:
    winner = winner(board)
    if winner is None:
      print(f'Game Over: Tie.')
    else:
      print(f'Game Over: {winner} wins.')
    break
  else:
    if user != playr and not game_over:
      if ai_turn:
        move = minimax(board)
        board = result(board, move)
        ai_turn = False
        print(numpy.array(board))
    elif user == playr and not game_over:
      ai_turn = True
      print('Enter the position to move (row,col)')
      i = int(input('Row:'))
      j = int(input('Col:'))
      if board[i][j] == EMPTY:
        board = result(board, (i, j))
        print(numpy.array(board))
```

<strong>Thuật toán <code>Alpha–Beta Pruning</code>:</strong>
Alpha–Beta Pruning là một cải tiến của thuật toán Minimax, giúp giảm số lượng trạng thái cần duyệt mà vẫn đảm bảo kết quả tối ưu. Thuật toán “cắt tỉa” những nhánh mà chắc chắn không ảnh hưởng đến quyết định cuối cùng.
<br><br><strong>Ý tưởng chính:</strong>

<strong>Alpha (α)</strong>: giá trị tốt nhất mà Max (người chơi tối đa) có thể đạt được từ các nhánh đã duyệt.

<strong>Beta (β)</strong>: giá trị tốt nhất mà Min (người chơi tối thiểu) có thể đạt được từ các nhánh đã duyệt.

Khi phát hiện một nhánh không thể cải thiện kết quả so với α hoặc β → cắt tỉa nhánh đó, không cần duyệt tiếp.

import các thư viện

```python
import os
import math
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>GetWinner(board)</code>:</strong>
Kiểm tra trạng thái kết thúc game bằng cách so sánh giá trị các ô trên 8 đường thắng (ngang, dọc, chéo) và trả về ký tự người thắng hoặc <code>None</code> nếu chưa có ai thắng.
</div>

```python
def GetWinner(board):
    """
    Returns the winner in the current board if there is one, otherwise it
    returns None.
    """
    # horizontal
    if board[0] == board[1] and board[1] == board[2]:
        return board[0]
    elif board[3] == board[4] and board[4] == board[5]:
        return board[3]
    elif board[6] == board[7] and board[7] == board[8]:
        return board[6]
    # vertical
    elif board[0] == board[3] and board[3] == board[6]:
        return board[0]
    elif board[1] == board[4] and board[4] == board[7]:
        return board[1]
    elif board[2] == board[5] and board[5] == board[8]:
        return board[2]
    # diagonal
    elif board[0] == board[4] and board[4] == board[8]:
        return board[0]
    elif board[2] == board[4] and board[4] == board[6]:
        return board[2]
    return None
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>PrintBoard(board)</code>:</strong>
Làm sạch màn hình console và hiển thị trạng thái hiện tại của bàn cờ dưới dạng lưới 3x3 giúp cập nhật giao diện cho người chơi.
</div>

```python
def PrintBoard(board):
    """
    Clears the console and prints the current board.
    """
    os.system('cls' if os.name == 'nt' else 'clear')
    print(f'''
    {board[0]}|{board[1]}|{board[2]}
    {board[3]}|{board[4]}|{board[5]}
    {board[6]}|{board[7]}|{board[8]}
    ''')
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>GetAvailableCells(board)</code>:</strong>
Trả về danh sách các ô còn trống trên bàn cờ. Hàm duyệt qua danh sách trạng thái và thu thập những vị trí chưa bị chiếm bởi 'X' hoặc 'O'
</div>

```python
def GetAvailableCells(board):
    """
    Returns a list of indices containing all available cells in a board.
    """
    available = list()
    for cell in board:
        if cell != "X" and cell != "O":
            available.append(cell)
    return available
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>minimax(position, depth, alpha, beta, isMaximizing)</code>:</strong>
Đánh giá giá trị của trạng thái bàn cờ hiện tại bằng thuật toán Minimax đệ quy có tối ưu hóa bằng cắt tỉa Alpha-Beta. Hàm trả về điểm số dự đoán của ván đấu từ góc nhìn của người chơi.

<br><strong>Cách hoạt động:</strong>

<ol>
<li><strong>Kiểm tra trạng thái kết thúc (Base Case):</strong>
    <ul>
        <li>Nếu có người thắng: Trả về <code>10 - depth</code> (nếu X thắng) hoặc <code>-10 + depth</code> (nếu O thắng). Việc trừ/cộng <code>depth</code> giúp AI ưu tiên chiến thắng ngắn nhất và trì hoãn thất bại lâu nhất.</li>
        <li>Nếu hòa (hết ô trống): Trả về <code>0</code>.</li>
    </ul>
</li>

<li><strong>Lượt Tối đa hóa (Maximizing - X):</strong>
    <ul>
        <li>Khởi tạo <code>maxEval</code> là <code>-∞</code>.</li>
        <li>Duyệt qua các ô trống, thử đi nước "X".</li>
        <li>Gọi đệ quy <code>minimax</code> cho lượt tiếp theo (của O).</li>
        <li>Cập nhật <code>maxEval</code> lớn nhất tìm được và cập nhật <code>alpha</code>.</li>
        <li><strong>Cắt tỉa:</strong> Nếu <code>beta <= alpha</code>, dừng vòng lặp (nhánh này không cần xét tiếp).</li>
    </ul>
</li>

<li><strong>Lượt Tối thiểu hóa (Minimizing - O):</strong>
    <ul>
        <li>Khởi tạo <code>minEval</code> là <code>+∞</code>.</li>
        <li>Duyệt qua các ô trống, thử đi nước "O".</li>
        <li>Gọi đệ quy <code>minimax</code> cho lượt tiếp theo (của X).</li>
        <li>Cập nhật <code>minEval</code> nhỏ nhất tìm được và cập nhật <code>beta</code>.</li>
        <li><strong>Cắt tỉa:</strong> Nếu <code>beta <= alpha</code>, dừng vòng lặp.</li>
    </ul>
</li>
</ol>

<strong>Giá trị trả về:</strong>
Một số nguyên (int) biểu thị điểm số đánh giá của bàn cờ (ví dụ: 10, -8, 0...).
</div>

```python
def minimax(position, depth, alpha, beta, isMaximizing):
    winner = GetWinner(position)
    if winner != None:
        return 10 - depth if winner == "X" else -10 + depth

    if len(GetAvailableCells(position)) == 0:
        return 0

    if isMaximizing:
        maxEval = -math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "X"
            Eval = minimax(position, depth + 1, alpha, beta, False)
            maxEval = max(maxEval, Eval)
            alpha = max(alpha, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return maxEval
    else:
        minEval = +math.inf
        for cell in GetAvailableCells(position):
            position[cell - 1] = "O"
            Eval = minimax(position, depth + 1, alpha, beta, True)
            minEval = min(minEval, Eval)
            beta = min(beta, Eval)
            position[cell - 1] = cell
            if beta <= alpha:
                break # prune
        return minEval
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Hàm <code>FindBestMove(currentPosition, AI)</code>:</strong>
Đóng vai trò là điểm khởi đầu cho lượt đi của AI. Hàm duyệt qua tất cả các ô trống, sử dụng hàm <code>minimax</code> để đánh giá tiềm năng của từng nước đi và chọn ra vị trí tối ưu nhất.

<br><strong>Cách hoạt động:</strong>
<ol>
<li>Khởi tạo giá trị tốt nhất (<code>bestVal</code>): <code>-∞</code> nếu AI là "X" (muốn tối đa hóa), hoặc <code>+∞</code> nếu AI là "O" (muốn tối thiểu hóa).</li>

<li>Duyệt qua từng ô trống (<code>cell</code>) trong danh sách <code>GetAvailableCells</code>:
    <ul>
        <li>Thực hiện nước đi thử: Đặt quân cờ của AI vào ô đó.</li>
        <li>Gọi <code>minimax(...)</code> để tính điểm số dự đoán cho trạng thái bàn cờ mới này.</li>
        <li>Hoàn tác nước đi (backtrack): Trả lại giá trị ô ban đầu để không ảnh hưởng đến các lần thử tiếp theo.</li>
    </ul>
</li>

<li>So sánh và cập nhật:
    <ul>
        <li>Nếu AI là "X" và điểm mới > <code>bestVal</code>: Cập nhật <code>bestVal</code> và lưu vị trí <code>bestMove</code>.</li>
        <li>Nếu AI là "O" và điểm mới < <code>bestVal</code>: Cập nhật <code>bestVal</code> và lưu vị trí <code>bestMove</code>.</li>
    </ul>
</li>

<li>Sau khi thử hết các ô, trả về <code>bestMove</code> — chỉ số của ô được chọn.</li>
</ol>

<strong>Giá trị trả về:</strong>
Một số nguyên (int) đại diện cho vị trí ô trên bàn cờ (ví dụ: 1-9) là nước đi tốt nhất.
</div>

```python
def FindBestMove(currentPosition, AI):
    bestVal = -math.inf if AI == "X" else +math.inf
    bestMove = -1
    for cell in GetAvailableCells(currentPosition):
        currentPosition[cell - 1] = AI
        moveVal = minimax(currentPosition, 0, -math.inf, +math.inf, False if AI == "X" else True)
        currentPosition[cell - 1] = cell

        if (AI == "X" and moveVal > bestVal):
            bestMove = cell
            bestVal = moveVal
        elif (AI == "O" and moveVal < bestVal):
            bestMove = cell
            bestVal = moveVal
    return bestMove
```

<div style="font-family: Arial, sans-serif; border: 2px solid #4CAF50; padding: 12px; border-radius: 8px; max-width: 650px;">

<strong>Vòng lặp chính:</strong>
Điều phối luồng đi của trò chơi, quản lý luân phiên lượt đi giữa Người và AI, đồng thời kiểm tra điều kiện kết thúc ván đấu.

<br><strong>Quy trình hoạt động:</strong>
<ol>
<li><strong>Khởi tạo:</strong>
    <ul>
        <li>Yêu cầu người chơi chọn phe ("X" hoặc "O").</li>
        <li>Tạo bàn cờ ban đầu là danh sách các số từ 1 đến 9.</li>
        <li>Thiết lập "X" đi trước.</li>
    </ul>
</li>

<li><strong>Vòng lặp game (`while True`):</strong>
    <ul>
        <li><strong>Lượt AI:</strong> Tự động tính toán nước đi tối ưu bằng <code>FindBestMove</code>, cập nhật bàn cờ và chuyển lượt.</li>
        <li><strong>Lượt Người chơi:</strong> Hiển thị bàn cờ, lặp lại việc yêu cầu nhập số (có xử lý lỗi ngoại lệ <code>ValueError</code>) cho đến khi nhận được nước đi hợp lệ.</li>
    </ul>
</li>

<li><strong>Kiểm tra kết thúc (sau mỗi lượt):</strong>
    <ul>
        <li>Nếu <code>GetWinner</code> trả về kết quả: Thông báo người thắng và kết thúc vòng lặp.</li>
        <li>Nếu không có người thắng nhưng đã đi đủ 9 nước (biến <code>counter</code>): Thông báo Hòa ("Tie") và kết thúc.</li>
    </ul>
</li>
</ol>
</div>

```python
player = input("Play as X or O? ").strip().upper()
AI = "O" if player == "X" else "X"
currentGame = [*range(1, 10)]
# X always starts first.
currentTurn = "X"
counter = 0

while True:
    if currentTurn == AI:
        # NOTE: if the AI starts first, it'll always choose index 0 so
        # to save time you could play it.
        cell = FindBestMove(currentGame, AI)
        currentGame[cell - 1] = AI
        currentTurn = player
    elif currentTurn == player:
        PrintBoard(currentGame)
        while True:
            try:
                humanInput = int(input("Enter Number: ").strip())
                if humanInput in currentGame:
                    currentGame[humanInput - 1] = player
                    currentTurn = AI
                    break
                else:
                    PrintBoard(currentGame)
                    print("Cell Not Available.")
            except ValueError:
                print("Invalid input. Please enter a number.")

    if GetWinner(currentGame) != None:
        PrintBoard(currentGame)
        print(f"{GetWinner(currentGame)} WON!!!")
        break

    counter += 1
    if GetWinner(currentGame) == None and counter == 9:
        PrintBoard(currentGame)
        print("Tie.")
        break
```
